```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df_muliple_choice_reponses = pd.read_csv("multipleChoiceResponses.csv",encoding = "ISO-8859-1")
df_freeform = pd.read_csv("freeformResponses.csv")
df_Rates = pd.read_csv("conversionRates.csv")
```

```python
# This Calculates the types of Status of the Employee
# employee_full_time,unemployee,freelancers = 0,0,0
a=np.array([0,0,0,0,0,0])
for i in range(len(df_muliple_choice_reponses)):
    if(df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'Employed full-time'):
        a[0] += 1
    elif (df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'Not employed, but looking for work'):
        a[1] +=1
    elif (df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'Independent contractor, freelancer, or self-employed'):
        a[2] +=1
    elif(df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'Not employed, and not looking for work'):
        a[3] +=1
    elif(df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'Employed part-time'):
        a[4] +=1
    elif(df_muliple_choice_reponses['EmploymentStatus'].get_value(i) == 'I prefer not to say'):
        a[5] +=1
    
        
```

```python
ls = []
for i in range(len(a)):
    ls.append(a[i])
    
ls
```

```python
data = ['Employed full-time','Not employed, but looking for work','Independent contractor, freelancer, or self-employed','Not employed, and not looking for work','Employed part-time','I prefer not to say']
f,ax=plt.subplots(figsize=(10,10))
EmployementStatus=plt.bar(data,ls,align = "center")
plt.xticks(rotation = 90)
EmployementStatus[0].set_color('g')
EmployementStatus[1].set_color('r')
EmployementStatus[2].set_color('b')
EmployementStatus[4].set_color('y')
EmployementStatus[5].set_color('k')
plt.title('Employee Status')
```

```python
df_muliple_choice_reponses['Country']
c_ar = np.array([0,0,0,0,0,0])
for i in range(len(df_muliple_choice_reponses)):
    if (df_muliple_choice_reponses['Country'].get_value(i) == "India"):
        c_ar[0] +=1
    elif(df_muliple_choice_reponses['Country'].get_value(i) == "Japan"):
        c_ar[1] +=1
    elif(df_muliple_choice_reponses['Country'].get_value(i) == "Australia"):
        c_ar[2] +=1
    elif(df_muliple_choice_reponses['Country'].get_value(i) == "United States"):
        c_ar[3] +=1
    elif(df_muliple_choice_reponses['Country'].get_value(i) == "Germany"):
        c_ar[4] +=1
    elif(df_muliple_choice_reponses['Country'].get_value(i) == "Russia"):
        c_ar[5] +=1
c_ar
c_ls = []
for i in range(len(c_ar)):
    c_ls.append(c_ar[i])
c_ls
labels = 'India','Japan','Australia','US','Germany','Russia'
explode = (0.3,0,0,0,0,0)
plt.pie(c_ls,labels = labels,explode = explode)
c_ls
```
